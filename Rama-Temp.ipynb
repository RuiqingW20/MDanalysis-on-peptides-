```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
import os
import glob
```

```python
import re
numbers = re.compile(r'(\d+)')
def numericalSort(value):
    parts = numbers.split(value)
    parts[1::2] = map(int, parts[1::2])
    return parts
```

```python
files = glob.glob('/Users/ruiqingwang/Downloads/Peptide simulation/Cypress/Rg-Rama-Temp/a-syn/Rama/*.xvg')
for file in sorted(files,key=numericalSort):
    data = pd.read_csv(file,skiprows = 34, header = None, sep = '\s+')
    data.columns = ['phi', 'psi', 'aa']
    df=data[data['phi'] <= 0]
    print(df)
```
