```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import re
numbers = re.compile(r'(\d+)')
def numericalSort(value):
    parts = numbers.split(value)
    parts[1::2] = map(int, parts[1::2])
    return parts
```

```python
import os
import glob
from os.path import basename,splitext
```

```python
files = glob.glob('nacl-rdf/*.xvg')

d=0

Rdf = []

for file in sorted(files,key=numericalSort):
    data = pd.read_csv(file, skiprows = 25, header = None, sep = '\s+')
    data.columns = ['r', 'gr']
    
    Rdf.append((data))


print(Rdf)

```

```python
font = {'family': 'Arial',
        'weight': 'normal',
        'size': 15,
        }
```

```python
fig, axs = plt.subplots(4, 4,figsize=(14, 15))
axs[0, 0].plot(Rdf[1].r,Rdf[1].gr)
axs[0, 1].plot(Rdf[2].r,Rdf[2].gr)
axs[0, 2].plot(Rdf[3].r,Rdf[3].gr)
axs[0, 3].plot(Rdf[4].r,Rdf[4].gr)
axs[1, 0].plot(Rdf[5].r,Rdf[5].gr)
axs[1, 1].plot(Rdf[6].r,Rdf[6].gr)
axs[1, 2].plot(Rdf[7].r,Rdf[7].gr)
axs[1, 3].plot(Rdf[8].r,Rdf[8].gr)
axs[2, 0].plot(Rdf[9].r,Rdf[9].gr)
axs[2, 1].plot(Rdf[10].r,Rdf[10].gr)
axs[2, 2].plot(Rdf[11].r,Rdf[11].gr)
axs[2, 3].plot(Rdf[12].r,Rdf[12].gr)
axs[3, 0].plot(Rdf[13].r,Rdf[13].gr)
axs[3, 1].plot(Rdf[14].r,Rdf[14].gr)
# axs[3, 1].plot(Rdf[4].r,Rdf[14].gr)
custom_xlim =(0,2)
custom_ylim =(0,2)
plt.setp(axs, xlim=custom_xlim, ylim=custom_ylim);
flat_ax = axs.flatten()
title = ["2D", "3V", "4F", "5M", "6K","7G","8L","9S","10K","11A","12K","13E","14G","15V","",""]
for n, ax in enumerate(flat_ax):
    ax.set_title(f'{title[n]}')
plt.suptitle('rdf of Cl- backbone Nitrogen \n (NaCl)', fontsize='15')
fig.tight_layout()
plt.savefig('nacl_rdf_all.png', dpi=300)
plt.show()
```
